```python
def color_corr_red(value):
    
    """
  Colors elements in a dateframe
  green if correlation less than 0.6 and red if
  greater than 0.6. Does not color NaN
  values.
  """
    
    if value < 0.6:
        color = 'green'
        
    elif value > -0.6:
        color = 'red'

    elif value > 0.6:
        color = 'red'
        
    elif value < -0.6:
        color = 'red'
        
    else:
        color = 'black'

    return 'color: %s' % color
```

```python
#correlation
corr = data.corr()
corr.style.applymap(color_corr_red) 
```

```python
#skewness
corr = data.skew().to_frame()
corr.columns = ['Skewness']
corr
```

## Graphical EDA

```python
#heatmap
corr = data.corr()
plt.figure(figsize=(12,10))
sns.heatmap(data=corr, annot=True)
```

```python
#most sold products
data['Lineitem name'] = data['Lineitem name'].str.lower()
ten_most_sold = pd.DataFrame(data['Lineitem name'].value_counts())[:10]
ten_most_sold['name'] = ten_most_sold.index
ten_most_sold.reset_index(inplace=True)
plt.figure(figsize=(8,6))
plt.title('10 Most Sold Items')
sns.barplot(x='Lineitem name', y='name', data=ten_most_sold)
plt.xlabel('no. of sales')
```

```python
#most popular vendors
data['Vendor'] = data['Vendor'].str.lower()
pop_vend = pd.DataFrame(data['Vendor'].value_counts())[:10]
pop_vend['name'] = pop_vend.index
pop_vend.reset_index(inplace=True)
plt.figure(figsize=(8,6))
plt.title('10 Most Popular Vendors')
sns.barplot(x='Vendor', y='name', data=pop_vend)
plt.xlabel('no. of sales')
```

```python
#creating totals table excluding the nan columns
totals = data[data['Total'].isnull()==False]
```

```python
totals['Paid at'].dropna(inplace=True)
```

```python
totals['Paid at'] = pd.to_datetime(totals['Paid at'], format='%m/%d/%y %H:%M')
```

```python
for date in totals['Paid at']:
#    if date.islower() == True:
#        continue
#    else:
        date = pd.to_datetime(date, format='%m/%d/%y %H:%M')
```

```python
totals.index = totals['Paid at']
```

```python
totals.drop('Paid at', axis=1, inplace=True)
```

```python
#grouping by month
g = totals.groupby(pd.Grouper(freq="M"))
```

```python
sums = g.sum()
means = g.mean()
```

```python
plt.figure(figsize=(12,10))
sums['Total'].plot(title='Total Revenue per Month')
plt.show()

#spike in fall/winter, July drop not accurate, only half of the month's data included
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,8))

sums['Refunded Amount'].plot(ax=ax1, title='Total Refunded Amount per Month')

means['Refunded Amount'].plot(ax=ax2, title='Mean Refunded Amount per Month', color='#4c6ca0')

```
