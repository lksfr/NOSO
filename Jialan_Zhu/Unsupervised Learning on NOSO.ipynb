# 1. Clustering

## USJH order unit

```python
#create a list of categorical feature for clustering which is named as included
excluded = ["Name", "Currency", "Subtotal", "Shipping", "Taxes", "Total", "Discount Amount"]

included = ["Discount Code","Financial Status", "Fulfillment Status", "Accepts Marketing", "Vendor"]

#tobe_modified ={
#    "Paid at": "to year, month, day, weekday",
#    "Discount Code": "binary, is.nan?"
#}
```

```python
import re
import string

df = usjh_order[included].copy()
#dummfy the categorical features
dummies = pd.get_dummies(df)
#create better format for dummified features
dummies.rename(
    columns=lambda s: re.sub("["+string.punctuation+"]", '_', s.strip()).replace(" ", "_") + " ", 
    inplace=True
)
```

```python
#change the '1's into word (dummified feature names), and the 0*col seems left the empty string or space? (not NA)
for col in dummies:
    dummies[col] = [i*col for i in dummies[col]]
```

```python
#the .sum(axis=1) add all the element in each row into one long sentence, and strip spaces at two ends of the sentence
#documents is a list of sentences, each sentence present an order
documents = [s.strip() for s in dummies.sum(axis=1)]
```

```python
#CountVectorizer model convert a collection of text documents into a matrix of token counts
#in this case, each token is actually a dummified feature! the number of tokens is equal to the number dummified features
from sklearn.feature_extraction.text import CountVectorizer

tf_vectorizer = CountVectorizer()
tf = tf_vectorizer.fit_transform(documents)
tf_feature_names = tf_vectorizer.get_feature_names()
```

```python
#here we set n_component=10, which means we set 10 topics for the documents (this is like
# set K for K-mean clustering)
#then the LDA algorithm will calculate the probability of topic the sentence belongs to depending on the distribution of the words in it
from sklearn.decomposition import NMF, LatentDirichletAllocation

n_components = 10

lda = LatentDirichletAllocation(n_components=n_components, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)

lda.fit(tf)
lda.transform(tf)
```

```python
#we labal the sentence (the order in this case) the group number depending on the highest probability
group = np.argmax(lda.transform(tf), axis=1)
```

```python
#we assume token word count matrix as new features, group number as target, run a lasso regression
#to determine the important features (or token word or dummified features in this case) decide the group distribution
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.006, normalize=True)
lasso.fit(tf, group)
# lasso.coef_
np.array(tf_feature_names)[lasso.coef_!=0]
```

```python
from sklearn.linear_model import LogisticRegression

logit = LogisticRegression(penalty="l1", C=0.1)
logit.fit(tf, group)
logit.coef_
```

```python
np.array(tf_feature_names)[logit.coef_[0]!=0]
```

```python
#input apriori project, return the association rule data frame
def associationR(df, target, groupby='name', min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2):
    '''print out model parameters'''
    print('Association rules of {} appearing in {}'.format(target,groupby))
    print('the minimum support of relations (support threshold): ', min_support)
    print('the minimum confidence of relations (confidence threshold): ', min_confidence)
    print('the minimum lift of relations: ', min_lift)
    print('the minimum number of item in the base item: ', min_length)
    
    '''preprocessing'''
    df=df.groupby(groupby)[target]
    
    lis=[]
    for key, value in df:
        lis.append(list(value))
    
    '''apply model'''
    association_rules = apriori(lis, 
                                min_support=min_support,
                                min_confidence=min_confidence, 
                                min_lift=min_lift, 
                                min_length=min_length)
    
    '''generate result data frame'''
    association_rules=list(association_rules)
    ar={}
    ar['item_base']=[]
    ar['item_add']=[]
    ar['support']=[]
    ar['confidence']=[]
    ar['lift']=[]

    for item in association_rules:

        item_base=[x for x in item[2][0][0]]
        ar['item_base'].append(item_base)

        item_add=[x for x in item[2][0][1]]
        ar['item_add'].append(item_add)

        support=item[1]
        ar['support'].append(support)

        confidence=item[2][0][2]
        ar['confidence'].append(confidence)

        lift=item[2][0][3]
        ar['lift'].append(lift)

    ar=pd.DataFrame(ar)

    return ar
```
